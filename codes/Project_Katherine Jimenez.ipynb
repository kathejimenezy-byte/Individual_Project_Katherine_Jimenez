## QUESTION 1: At the country level, how strongly are GDP per capita and CO₂ per capita related, and does this relationship change over time? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv("owid-co2-data.csv")

# 2. (Optional but good): keep only real countries (iso_code has 3 letters)
df = df[df['iso_code'].notna() & (df['iso_code'].str.len() == 3)]

# 3. Create GDP per capita
df['gdp_per_capita'] = df['gdp'] / df['population']

# 4. Keep only needed columns and drop missing values
df_small = df[['country', 'year', 'gdp_per_capita', 'co2_per_capita']].dropna()

# 5. Correlation between GDP per capita and CO2 per capita *for each year*
corr_by_year = df_small.groupby('year').apply(
    lambda x: x['gdp_per_capita'].corr(x['co2_per_capita'])
)

# 6. Plot correlation over time
plt.figure(figsize=(8, 6))
plt.plot(corr_by_year.index, corr_by_year.values, linewidth = 3, color ='green')
plt.title("Correlation: GDP per Capita vs CO₂ per Capita (by year)", fontsize =14)
plt.xlabel("Year", fontsize =18)
# plt.xlim([1990, 1995])
plt.ylabel("Pearson correlation", fontsize =18)
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)
plt.savefig('q1_all_year.png', bbox_inches='tight', dpi =300)
plt.show()

# 7. Scatter plot for one example year
year_to_plot = 1991
df_year = df_small[df_small['year'] == year_to_plot]

plt.figure(figsize=(8, 6))
plt.scatter(df_year['gdp_per_capita'], df_year['co2_per_capita'])

# Add annotations
for _, row in df_year.iterrows():
    plt.annotate(
        row['country'],
        (row['gdp_per_capita'], row['co2_per_capita']),
        fontsize=8,
        alpha=1.0
    )

plt.title(f"GDP per Capita vs CO₂ per Capita ({year_to_plot})", fontsize = 14)
plt.xlabel("GDP per capita", fontsize =18)
plt.ylabel("CO₂ per capita", fontsize =18)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=14) 
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('q1_1991.png', bbox_inches='tight', dpi =300)
plt.show()

```

## QUESTION 2: Can we predict the CO₂ per capita from GDP per capita?

### First attempt: linear regresion

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load data
df = pd.read_csv("owid-co2-data.csv")
df = df[df['iso_code'].notna() & (df['iso_code'].str.len() == 3)]
df['gdp_per_capita'] = df['gdp'] / df['population']

# Choose the year you want
year_to_plot = 2010

df_year = df[df['year'] == year_to_plot][['country', 'gdp_per_capita', 'co2_per_capita']].dropna()

# Extract X and y for regression
X = df_year[['gdp_per_capita']].values     # must be 2D
y = df_year['co2_per_capita'].values       # 1D

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Regression line values
x_line = np.linspace(df_year['gdp_per_capita'].min(),
                     df_year['gdp_per_capita'].max(), 200)
y_line = model.predict(x_line.reshape(-1, 1))

# Plot scatter + regression line
plt.figure(figsize=(7, 6))
plt.scatter(df_year['gdp_per_capita'], df_year['co2_per_capita'], label="Countries", alpha=1.0)
plt.plot(x_line, y_line, color='red', linewidth=4, label="Linear regression")

plt.title(f"GDP per Capita vs CO₂ per Capita ({year_to_plot}) — Linear regression", fontsize = 12)
plt.xlabel("GDP per capita", fontsize = 18)
plt.ylabel("CO₂ per capita", fontsize = 18)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16) 
plt.legend(fontsize = 14)
# Add R^2 annotation
r2 = model.score(X, y)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q2_linear.png', bbox_inches='tight', dpi =300)
plt.show()

# Print coefficients
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("R²:", model.score(X, y))
```

### Second attempt: Quadratic model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv("owid-co2-data.csv")
df = df[df['iso_code'].notna() & (df['iso_code'].str.len() == 3)]
df['gdp_per_capita'] = df['gdp'] / df['population']

# Pick your year
year_to_plot = 2010
df_year = df[df['year'] == year_to_plot][['gdp_per_capita','co2_per_capita']].dropna()

# Extract x and y
x = df_year['gdp_per_capita'].values.reshape(-1, 1)
y = df_year['co2_per_capita'].values.reshape(-1, 1)

# Create quadratic features: [x, x^2]
X_quad = np.hstack([x, x**2])

# Fit model
model = LinearRegression()
model.fit(X_quad, y)

# Smooth curve for plotting
x_line = np.linspace(x.min(), x.max(), 300).reshape(-1, 1)
X_line_quad = np.hstack([x_line, x_line**2])
y_line = model.predict(X_line_quad)

# Plot scatter + curve
plt.figure(figsize=(7, 6))
plt.scatter(x, y, label="Countries", alpha=1.0)
plt.plot(x_line, y_line, color="red", linewidth=4, label="Quadratic fit")
plt.title(f"GDP per Capita vs CO₂ per Capita ({year_to_plot}) — Quadratic Model", fontsize = 12)
plt.xlabel("GDP per capita",fontsize = 18)
plt.ylabel("CO₂ per capita",fontsize = 18)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16) 
plt.legend(fontsize =14)
# Add R^2 annotation
r2 = model.score(X_quad, y)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q2_quad.png', bbox_inches='tight', dpi =300)
plt.show()

# Print coefficients
a = model.coef_[0][1]   # coefficient of x^2
b = model.coef_[0][0]   # coefficient of x
c = model.intercept_[0]

print("Model:  y = a x^2 + b x + c")
print("a =", a)
print("b =", b)
print("c =", c)
print("R² =", model.score(X_quad, y))
```

## Power-law fitting and prediction

### Year 2010

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Load data
df = pd.read_csv("owid-co2-data.csv")
df = df[df['iso_code'].notna() & (df['iso_code'].str.len() == 3)]
df['gdp_per_capita'] = df['gdp'] / df['population']

# Choose year
year_to_plot = 2010
df_year = df[df['year'] == year_to_plot][['gdp_per_capita','co2_per_capita']].dropna()

x = df_year['gdp_per_capita'].values
y = df_year['co2_per_capita'].values

# Remove zeros
mask = (x > 0) & (y > 0)
x = x[mask]
y = y[mask]

# Log-space
logx = np.log(x).reshape(-1, 1)
logy = np.log(y).reshape(-1, 1)

# ========== TRAIN / TEST SPLIT ==========
logx_train, logx_test, logy_train, logy_test = train_test_split(
    logx, logy, test_size=0.2, random_state=42
)

# Fit model
model = LinearRegression()
model.fit(logx_train, logy_train)

# Extract power-law parameters
p = model.coef_[0][0]
a = np.exp(model.intercept_[0])

# Smooth fit curve
x_line = np.linspace(x.min(), x.max(), 300)
y_line = a * (x_line ** p)

# Convert back to normal space for plotting train/test
x_train_original = np.exp(logx_train).flatten()
y_train_original = np.exp(logy_train).flatten()

x_test_original = np.exp(logx_test).flatten()
y_test_original = np.exp(logy_test).flatten()

# ========== PLOT ==========
plt.figure(figsize=(7,6))

# Train points (red)
plt.scatter(x_train_original, y_train_original, 
            color="red", label="Train", alpha=1.0)

# Test points (blue)
plt.scatter(x_test_original, y_test_original, 
            color="blue", label="Test", alpha=1.0)

# Fitted curve (black)
plt.plot(x_line, y_line, color="green", linewidth=4, label="Power-law fit")

plt.xlabel("GDP per capita", fontsize = 18)
plt.ylabel("CO₂ per capita", fontsize = 18)
plt.xscale('log')
plt.yscale('log')
plt.xticks(fontsize=16) 
plt.yticks(fontsize=16) 
plt.title(f"Power-Law Fit $y = a x^p$ (Train=Red, Test=Blue) — {year_to_plot}", fontsize =14)
plt.xlabel("GDP per capita")
plt.ylabel("CO₂ per capita")
plt.legend(fontsize = 16)
# Add R^2 annotation
r2 =  model.score(logx, logy)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q2_fit_and_pred_2010.png', bbox_inches='tight', dpi =300)
plt.show()

# ========== METRICS ==========
print("Power-law model:  y = a x^p")
print("a =", a)
print("p =", p)
print("Train R²:", r2_score(logy_train, model.predict(logx_train)))
print("Test R² :", r2_score(logy_test,  model.predict(logx_test)))
```

## QUESTION 3: Do countries exhibit evidence for an Environmental Kuznets Curve (EKC)?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv("owid-co2-data.csv")
df = df[df['iso_code'].notna() & (df['iso_code'].str.len() == 3)]
df['gdp_per_capita'] = df['gdp'] / df['population']

# Pick the country you want
country = "United States"   # ← change this to test another country

df_country = df[df['country'] == country][['year','gdp_per_capita','co2_per_capita']].dropna()

# Extract x and y
x = df_country['gdp_per_capita'].values.reshape(-1, 1)
y = df_country['co2_per_capita'].values.reshape(-1, 1)

# Quadratic features
X_quad = np.hstack([x, x**2])

# Fit model y = a + b x + c x^2
model = LinearRegression()
model.fit(X_quad, y)

a = model.intercept_[0]
b = model.coef_[0][0]
c = model.coef_[0][1]

print("Country:", country)
print("a =", a)
print("b =", b)
print("c =", c)
print("R² =", model.score(X_quad, y))

# Turning point (if EKC exists)
if c < 0:
    turning_point = -b / (2*c)
    print("EKC turning point GDP per capita =", turning_point)
else:
    print("No EKC (no inverted-U shape)")

# Plot curve
x_line = np.linspace(x.min(), x.max(), 300).reshape(-1, 1)
y_line = model.predict(np.hstack([x_line, x_line**2]))

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='#FF8C00', label="Historical data")
plt.plot(x_line, y_line, color="purple", linewidth=4, label="Quadratic fit")
plt.title(f"EKC Test for {country}", fontsize =20)
plt.xlabel("GDP per capita", fontsize =20)
plt.ylabel("CO₂ per capita", fontsize =20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize = 15)
# Add R^2 annotation
r2 =  model.score(X_quad, y)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q3_EKC_USA.png', bbox_inches='tight', dpi =300)
plt.savefig("name.png", bbox_inches='tight', dpi=300)
plt.show()
```

## QUESTION 4: Does global CO₂ help predict global temperature anomalies? 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# ==============================
# 1. Load GLOBAL CO2 (OWID)
# ==============================
co2 = pd.read_csv("owid-co2-data.csv")

# Keep only the global aggregate ("World")
co2_world = co2[co2["country"] == "World"][["year", "co2"]].rename(
    columns={"co2": "co2_global"}
)

# Optional: focus on modern era
co2_world = co2_world[co2_world["year"] >= 1960]

# ==============================
# 2. Load NASA temperature data
# ==============================
# First row in GLB.Ts+dSST.csv is metadata, so skip it
temp = pd.read_csv("GLB.Ts+dSST.csv", skiprows=1)

# Keep Year and annual mean (J-D)
temp = temp[["Year", "J-D"]].rename(columns={
    "Year": "year",
    "J-D": "temp_anomaly"
})

# Convert to numeric (some early values may be '***')
temp["temp_anomaly"] = pd.to_numeric(temp["temp_anomaly"], errors="coerce")

# Same period and drop missing
temp = temp[temp["year"] >= 1960].dropna(subset=["temp_anomaly"])

# ==============================
# 3. Merge CO2 + temperature
# ==============================
df = pd.merge(co2_world, temp, on="year", how="inner").dropna()

X = df[["co2_global"]]        # feature: global CO2 emissions
y = df["temp_anomaly"]        # target: global temp anomaly

# ==============================
# 4. Train/Test split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=True,
    random_state=42
)

# ==============================
# 5. Fit linear regression
# ==============================
model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print("R² (train):", r2_score(y_train, y_train_pred))
print("R² (test): ", r2_score(y_test, y_test_pred))
print("RMSE (test):", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("Slope (β):", model.coef_[0])
print("Intercept (α):", model.intercept_)

# ==============================
# 6. PLOT 1: Training data + best-fit curve
#    x = true CO2 (train), y = true temp anomaly (train)
# ==============================
# Sort training data by CO2 for a smooth line
X_train_sorted = X_train.sort_values(by="co2_global")
y_train_sorted = y_train.loc[X_train_sorted.index]
y_line = model.predict(X_train_sorted)

plt.figure(figsize=(7, 6))
plt.scatter(X_train["co2_global"], y_train, alpha=1.0, label="Training data", color = 'red')
plt.plot(X_train_sorted["co2_global"], y_line,
         color="green", linewidth=4, label="Best-fit line")

plt.xlabel("Global CO₂ emissions", fontsize =18)
plt.ylabel("Temperature anomaly (°C)", fontsize  =18)
plt.title("TRAINING SET: Temp anomaly vs CO₂\nwith best-fit linear curve", fontsize =18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize = 16)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q4_temp_vs_CO2.png', bbox_inches='tight', dpi =300)
plt.show()
print("\n\n")
# ==============================
# 7. PLOT 2: Test true vs predicted temp anomaly
#    x = true temp anomaly (test), y = predicted temp anomaly (test)
# ==============================
plt.figure(figsize=(7, 6))
plt.scatter(y_test, y_test_pred, alpha=1.0, label="Test points", color ='blue')

# 1:1 reference line
min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val],
         color="black", linewidth=4, label="y = x (perfect prediction)")

plt.xlabel("True temperature anomaly (°C)", fontsize =18)
plt.ylabel("Predicted temperature anomaly (°C)", fontsize =18)
plt.title("TEST SET: True vs Predicted \n Temperature Anomaly", fontsize =18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize = 16)
plt.text(
    0.05, 0.95,
    f"$R^2 = {r2:.2f}$",
    transform=plt.gca().transAxes,
    fontsize=16,
    verticalalignment='top'
)
plt.tight_layout()
plt.savefig('q4_true_vs_pred.png', bbox_inches='tight', dpi =300)
plt.show()
```
